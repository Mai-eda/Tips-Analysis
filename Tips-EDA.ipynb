# Countplot of Tables served by Day

```python
o=sns.countplot(x='day',data=df,palette=['skyblue','pink','lavender','lightcyan'],order=day_order)
sns.set(style='whitegrid')
plt.title('The Count of Table Served by Day')
```

```python
sns.set(style='darkgrid')
# HORIZONTAL BAR
sns.countplot(y='sex',hue='smoker',data=df,palette=['pink','skyblue'])
```

```python
sns.histplot(data=df,x='total_bill',stat='count',kde=True)
```

```python
# BOX PLOT 

sns.boxplot(x='day', y='total_bill',data=df, hue='smoker', color='grey')
```

```python
#PAIR PLOT

df.loc[:, ['total_bill','tip','sex','smoker','size']];
gender_pal = sns.color_palette("Set2");

sns.pairplot(df.loc[:, ['total_bill','tip','sex','smoker','size']], hue="sex",palette=gender_pal);
```
